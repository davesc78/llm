```python
import torch
from transformers import pipeline
from google.colab import userdata
```

```python
model_name = "meta-llama/Meta-Llama-3.1-8B-Instruct"
torch.cuda.empty_cache()

pipe = pipeline(
    "text-generation",
    model=model_name,
        model_kwargs={
        "torch_dtype": torch.bfloat16,
        "quantization_config": {"load_in_4bit": True},
        "device_map": "cuda",
        "max_length": 256,  # Limit the maximum output length
        "pad_token_id": 128001,  # Set the padding token ID
    },
    batch_size=1
)



```

```python
def ask_llama (question, n_tokens = 256):
      messages = [
          {"role": "system", "content": 'you are an asistant. Limit your responses to {n_tokens} tokens'.format(n_tokens=n_tokens)},
          {"role": "user", "content": question},
      ]
      outputs = pipe(
          messages,
          max_new_tokens=256,
          pad_token_id=128001,
          #do_sample=False
      )
      assistant_response = outputs[0]["generated_text"][-1]["content"]
      return assistant_response
```

```python
print(ask_llama("¿Qué es Chilecompra?"))
```
